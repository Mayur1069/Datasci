```python
values = [1,2,3,4,5,6,9,10,12,55,60]

# From 1 - 12 the values are very close which can be said that 
# 90% of data is from 1 -12 and then 55 and 60 ,here 55, 60 are called outliers
```

```python
values 
```

```python
# Q1 is 25% of the data
# Q3 is 75% of the data

Q1= np.percentile(values,25)
```
