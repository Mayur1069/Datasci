### Measure of dispersion


```python
#variance 
# tells us abt : how is data spread around mean ?
# sigma^2 : variance of population
# s^2 : variance of sample


#  s^2 = summation(xi - xbar)^2 / n-1
# xi = elements of data xbar = mean

sample = [4,3,6,5,2]
xbar = 4
result = 0
for i in sample :
    s = ((i - xbar)**2)
    result += s
variance = result/(len(sample)-1)
print ( variance)

#o/p = 2.5
#this 2.5 is near xbar so it's said the data is nicely spread
#helps to find status of outliers

#standard deviation
# becasue of squared values variance gives extra weighr to outliers
# stand. devaitin = variance ** 1/2 (square root)

# sigma : standard deviation of population 
# S : standard deviation of sample 
# Large Standard Deviation = Large Spread
print (f"standard Deviation : {variance ** (1/2)}" )
```

```python
#POPULATION VARIANCE  /  SAMPLE VARIANCE
import numpy as np
display.Image(r"C:\Users\max\Desktop\code practice\popvar_sampvar.png")
```

```python
#POPULATION VARIANCE

population = [5,10,15,220,25,300,35,40,45]
```

```python
pop_mean = sum(population)/ len(population) 
pop_mean
```

```python
sq_pop_deviation = [(i - pop_mean)**2 for i in population]
```

```python
sq_pop_deviation
```

```python
pop_variance = sum(sq_pop_deviation)/len(population) 
pop_variance
```

```python
#the data is widely spread.
```

```python
# SAMPLE VARIANCE

sample = np.arange(5,25,5)

#
# In NumPy, np.arange is a function 
# that creates an array with 
# regularly spaced values within a specified range. 
# The syntax for np.arange is as follows:

#  numpy.arange([start, ]stop, [step, ])
```

```python
sample
```

```python
s_mean = sum(sample)/len(sample)
s_mean
```

```python
sq_s_deviation = [(i-s_mean)**2 for i in sample]
sq_s_deviation
```

```python
s_variance = sum(sq_s_deviation)/len(sample)-1
s_variance
```
