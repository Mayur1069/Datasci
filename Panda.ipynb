# Panda 

series and dataframes

Iamagine a table having rows and columns of data.
the column is Series and combination of the series is called Dataframe/Dataset

1 col + multiple rows  = series

multiple rows + mulitple col = dataframe/dataset

```python
#Panda series is like a column in a table 
#It is a 1 Dimensional array holding data of any type.

import pandas as pd

series = pd.Series((1,7,2,8,5))

print (series)
```

```python
#series Method with Index

series = pd.Series((1,7,2,8,5) , index =['a','b','c','d','e'] )

print(series)
```

```python
#Accessing Series elements

print(series['a'])
print(series['d'])
print(series[0])
```

#### Pandas DataFrame

```python
#DataFrame Method on list

data = pd.DataFrame([1,2,3,4,5])
print(data)
```

```python
#DataFrame using Dictionary

data = pd.DataFrame({'Name' :['tom','jack','steve','ricky','sam','will','root'],
                       'Age':[28,35,25,24,26,27,34]} )
data.head(10)
```

```python
data.head(5)
```

```python
data = pd.DataFrame({'Name' :['tom','jack','steve','ricky','sam','will','root'],
                       'Age':[28,35,25,24,26,27,34]}, index = ['c1','c2','c3','c4','c5','c6','c7'] )

data.head(7)
```

```python
#Column Selection

print(data['Name'])
print(data['Age'])
```

```python
#row selection

data.loc['c1','Age']  #onlylocation
```

```python
data.iloc[5,1]  #integerlocation [index (row) , column number]
```

### Pandas concantenation and indexing

```python
names = pd.DataFrame({'Name' :['tom','jack','steve','ricky','sam','will','root']})
ages = pd.DataFrame({'Age' : [28,35,25,24,26,27,34]} )
```

```python
names2 = pd.DataFrame({'Name2' : ['joe', 'papa', 'butter','james','cook']})
```

```python
pd.concat([names,ages], axis = 1)
```

```python
pd.concat([names,names2,ages],axis =1)
```

```python
pd.concat([names,names2])
```

Apply method

```python
data['marks'] = [201,210,195,183,225,203,190]
```

```python
data
```

```python
data['marks']=data['marks'].apply(lambda x: x/5)
```

```python
data
```

```python
#sort index

data.index = [10,28,64,7,1,6,15]   #assigning index

print(data)
```

```python
data.sort_index(ascending = True)        #Default Stateent : Ascending = true
```

```python
# for descending just turn the true value to false
data.sort_index(ascending = False) #this is aslo temporary
```

sorting can be done only on numerical data

```python
#another method for sorting

data.sort_values(by = 'marks',ascending = False)
```
