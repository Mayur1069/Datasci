# TASK 2

VIEW DISTRIBUTIONS OF THE VARIOUS FEATURES IN DATA SET AND CALCULATE THEIR CENTRAL TENDANCIES

WE WILL NOW LOOK AT THE DISTRIBUTION OF THE VARIOUS FEATURES IN THE DATASET

WE WILL ALSO CALCULATE APPROPRIATE MEASURES OF CENTRAL TENDANCY FOR THESE FEATURES

```python
#CREATE A HISTOGRAM OF 'FIXED ACIDITY'

plt.figure(figure = (11,6))

sns.histplot(data = data, x= 'fixed acidity', color = 'orange',alpha = .5, bins = 5)

plt.title("Histogram of fixed Acidity")
plt.xlabel("fixed Acidity")
plt.ylabel("Count")
plt.show()
```

```python
#obs 
#maximum number of wine unit has acidity ranging from 6 to 8
#histogram is normally distributed
```

```python
#lets see the measures of central tendancy working
# mean  = avg
# median  = central number
# mode = most repeated number
```

```python
#calculate the mean of fixed acidity

round(data['fixed acidity'].mean(),2)
```

```python
#calculate medain
data['fixed acidity'].median()
```

```python
#create histogram of 'fixed acidity' and also show mean and median
plt.figure(figure = (11,6))

sns.histplot(data = data, x= 'fixed acidity', color = 'orange',alpha = .5, bins = 5)

plt.title("Histogram of fixed Acidity")
plt.xlabel("fixed Acidity")
plt.ylabel("Count")
plt.vlines((data['fixed acidity'].mean()),ymin = 0 ,ymax = 4000 , color = 'blue',label = 'mean')
plt.vlines((data['fixed acidity'].median()),ymin = 0 ,ymax = 4000 , color = 'red',label = 'median')
plt.legend()
plt.show()
```

</n> Observation </p>
we can see that mean median are clear representative of data

mean and median are very close to each other to each other

we can choose eiher of the parameters say mean as the measure of central tendancy

```python
# Create displot using 'Volatile acidity' feature

plt.figure(figure = (11,6))

sns.distplot(data['volatile acidity'], color = 'green')

plt.title("displot of volatile Acidity")
plt.xlabel("volatile Acidity")
plt.ylabel("density")
plt.show()
```

the plot shows normal distribution


the normal distribution is described by the mean and the stnd. deviation

The normal distribution is often referred to as a 'bell curve' because of its shape:

1. the median and mean are equal
2. it has only one mode
3. it is symmetric, meaning it decreases the same amount on the left and the right of the centre

we can calculate skewness too using skew() function in python.

```python
#calculate skewness

data['volatile acidity'].skew()
```

```python
#we cas see the skewness(s) i greater than 1, hence it is positively skewwed
```
